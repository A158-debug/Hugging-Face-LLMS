```python
import os
import io
import gradio as gr
from transformers import pipeline
```

## Text Summarizer

```python

TextsummarizeModel = pipeline("summarization", model="facebook/bart-large-cnn",framework="pt", revision="main",use_auth_token=hf_api_key)

def Textsummarize(input):
    output = TextsummarizeModel(input)
    return output[0]['summary_text']
```

```python
text = ('''The tower is 324 metres (1,063 ft) tall, about the same height
        as an 81-storey building, and the tallest structure in Paris.
        Its base is square, measuring 125 metres (410 ft) on each side.
        During its construction, the Eiffel Tower surpassed the Washington
        Monument to become the tallest man-made structure in the world,
        a title it held for 41 years until the Chrysler Building
        in New York City was finished in 1930. It was the first structure
        to reach a height of 300 metres. Due to the addition of a broadcasting
        aerial at the top of the tower in 1957, it is now taller than the
        Chrysler Building by 5.2 metres (17 ft). Excluding transmitters, the
        Eiffel Tower is the second tallest free-standing structure in France
        after the Millau Viaduct.''')

print(Textsummarize(text))
```

```python

def Textsummarize(input):
    output = TextsummarizeModel(input)
    return output[0]['summary_text']

gr.close_all()
demo = gr.Interface(fn=Textsummarize, inputs=[gr.Textbox(label="Text to summarize", lines=6)], outputs=[gr.Textbox(label="Result", lines=3)])
demo.launch(debug=True,share=True)
```

## OpenAI GPT-3.5 with Gradio

```python
import openai
```

```python
def openAIModelGPT(input):
  openai.api_key = os.getenv("OPENAI_API_KEY")
  response = openai.Completion.create(
    model="text-davinci-003",
    prompt="/*{}*/".format(input),
    temperature=0,
    max_tokens=2000,
    top_p=1.0,
    frequency_penalty=0.0,
    presence_penalty=0.0,
    stop=["\"\"\""]
)
  if 'choices' in response:
    if len(response['choices']) > 0:
        return response['choices'][0]['text']
```

```python
def openAISummarizer(input):
    output = openAIModelGPT(input)
    return output

gr.close_all()
demo = gr.Interface(fn=openAISummarizer, inputs=[gr.Textbox(label="OpenAI gpt-3.5-turbo", lines=6)], outputs=[gr.Textbox(label="Result", lines=3)])
demo.launch(debug=True,share=True)
```

```python

ImageCaptioningModel = pipeline("image-to-text",model="Salesforce/blip-image-captioning-base")

def ImageCaptioning(input):
    output = ImageCaptioningModel(input)
    return output[0]['generated_text']
```
